```python
from pyspark import SparkContext,SparkConf
from pyspark.sql import SparkSession
# run spark master in all aviable cores-> 4 in this case
conf = pyspark.SparkConf().setAppName("SparkApp").setMaster('local[*]')
sc = pyspark.SparkContext(conf=conf) # sc is a spark context that creats an RDD object 
spark = SparkSession(sc)
```

```python
# importing pacakages 
import time
import collections
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

# count the number of words in movie title
### file size of 483 KB

```python
start = time.time()
# import the data and extract only the title column
def parseLine(line):
    fields = line.split(',') # split all the columns by the delimeter
    title = str(fields[1])  # grab only the title column of the movies
    title = re.compile(r'\W+', re.UNICODE).split(title.lower())   
    return (title)
lines = sc.textFile("C:/Users/afoto001.000/Desktop/Abraham_edu_doc/UvA/Master/Big data/BigData/bigdata/data/ml-latest-small/movies.csv")
rdd = lines.flatMap(parseLine)
```

```python
type(rdd)
```

```python
# a list of all in the movie title
words = rdd.flatMap(lambda x:x.split())
# word count in the list of rdd words
wordCounts = words.map(lambda x: (x, 1)).reduceByKey(lambda x, y: x + y)
# first interchange key by vcalue and then sort by key ( number of words)
wordCountsSorted = wordCounts.map(lambda x: (x[1], x[0])).sortByKey()
# collect all wordCountsSorted from the distribution (from the cores(4))
results = wordCountsSorted.collect()
```

```python
for result in results:
    count = str(result[0])
    word = result[1].encode('ascii', 'ignore')
    if (word):
        print(word.decode() + ":\t\t" + count)
end = time.time()
print("Time taken in seconds: " + str(end-start) + ' s')
```
