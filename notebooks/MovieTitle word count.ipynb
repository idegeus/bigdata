```python
from pyspark import SparkContext,SparkConf
from pyspark.sql import SparkSession
conf = pyspark.SparkConf().setAppName("SparkApp").setMaster('local[*]')
sc = pyspark.SparkContext(conf=conf) # sc is a spark context that creats an RDD object 
spark = SparkSession(sc)
```

```python
# importing pacakages 
import time
import collections
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

# count the number of words in movie title

```python
start = time.time()
# import the data and first collect all the words in a list
def parseLine(line):
    fields = line.split(',')
    title = str(fields[1])
    title = re.compile(r'\W+', re.UNICODE).split(title.lower())
    
    return (title)
lines = sc.textFile("C:/Users/afoto001.000/Desktop/Abraham_edu_doc/UvA/Master/Big data/BigData/bigdata/data/ml-latest-small/movies.csv")
rdd = lines.flatMap(parseLine)
```

```python
wordCounts = rdd.countByValue()
```

```python
for word, count in wordCounts.items():
    cleanWord = word.encode('ascii', 'ignore')    
    if (cleanWord):
        print(cleanWord.decode() + " " + str(count))
end = time.time()
print("Time taken in seconds: " + str(end-start) + ' s')
```
